Lo que logramos con los pasos anteriores en n8n fue probar que el flujo funcionaba. Ahora, nuestro objetivo es que el flujo sea continuo y automático. Para esto, debemos realizar dos modificaciones: una en n8n y otra en el código del ESP32.

Cambio a URL de Producción: Dentro del segundo workflow que creamos en n8n, entraremos al nodo de Webhook. En el encabezado superior, seleccionaremos la pestaña Production URL. Esta es la dirección que usaremos en el código del ESP32.

Nota importante: Mi URL original era http://raspberrypi.local:5678/webhook/lecturas-camara, pero para el ESP32 la cambié a http://192.168.100.53:5678/webhook/lecturas-camara. Esto es necesario porque el ESP32 no siempre reconoce el dominio .local y funciona de forma mucho más estable usando la IP directa de la Raspberry Pi. Guarda esta IP para el siguiente paso en el código del esp32.

Ajuste en el nodo de PostgreSQL (Evitar errores): Un cambio fundamental que debí hacer (ya que me arrojaba error) fue en el nodo de PostgreSQL que inserta los datos. Es necesario quitar los parámetros de id y fecha en el mapeo del nodo. Si intentas enviarlos manualmente o los dejas configurados sin datos, la base de datos puede detectar valores repetidos o nulos donde no corresponden y bloquear la subida. Al quitarlos, dejamos que la base de datos los genere automáticamente sin conflictos.

Activar el Workflow (Modo Producción): Con esta información lista, presionaremos el botón Publish (arriba a la derecha). asígnale un nombre que te acomode y actívalo. Al hacer esto, el flujo pasa de "Modo Test" a "Modo Producción", lo que significa que n8n estará escuchando continuamente la llegada de datos. Ya no es necesario presionar el botón Execute Workflow manualmente.
